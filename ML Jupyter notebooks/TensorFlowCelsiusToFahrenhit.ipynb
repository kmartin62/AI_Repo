```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

The formula for calculating Celsius to Fahrenheit is f = celsius * 1.8 + 32

```python
celsius_q = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
fahrenheit_q = np.array([-40, -14, 32, 46, 59, 72, 100], dtype=float)

for i, c in enumerate(celsius_q):
    print(f'{c} degrees Celsius is {fahrenheit_q[i]} degrees Fahrenheit')
```

```python
l0 = tf.keras.layers.Dense(units=4, input_shape=[1])
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=1)
model = tf.keras.Sequential([l0, l1, l2])
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(celsius_q, fahrenheit_q, epochs=500, verbose=False)
print("Finished training")
```

```python
history.history
```

```python
plt.xlabel('Epoch Number')
plt.ylabel('Loss Magnitude')
plt.plot(history.history['loss'])
```

```python
print(model.predict([100.0]))
```
